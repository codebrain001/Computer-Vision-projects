## Importing relevant libraries


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### read the test image

```python
image = cv2.imread("test_image.jpeg")
```

```python
print(image)
```

```python
image.shape
```

### To view the picture, we would have just use openCV but we have to create a special window, so we use 
### Matplotlib library to view the image in the current window.

- Matplotlib default color space is RGB
- OpenCV default color space is BGR

```python
# to view the image with matplotlib we use
plt.imshow(image)
```

```python
## the image is small
plt.figure(figsize=(15,15))
plt.imshow(image)
```

```python
### from the above cell, it was noted about the default color spaces, so we have to maintain constientence
## So, we have to convert from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(image)
```

```python
image_shape = image.shape
print(image_shape)
```

Image has shape (1080, 1920, 3). This is interpreted as;
- height = 1080
- width = 1920
- channels = 3

The 3 channels corresponds to the RGB values

```python
#lets check the value of the pixel in (300,300)
position = (300,300)
pixel_value = image[position]
print("The pixel value at position {} is {}".format(position, pixel_value))
```

#### We got 3 values for the pixel values at position 300,300 i.e height and width
#### 161 -> Red channel
#### 29  -> Green channel
#### 253 -> Blue channel

Here, the B-channel has the highest value so this position has a color close to blue

### Let's move to HSV

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(image)
```

## Let's go to bnw (GreyScale)


```python
#We have to convert the hsv to RGB first to gain original status
image  = cv2.cvtColor(image, cv2.COLOR_HSV2RGB)
# Then convert to Greyscale
image = cv2.cvtColor(image, cv2.COLOR_RGB)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(image)
```
